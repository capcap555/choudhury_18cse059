```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
path="https://raw.githubusercontent.com/capcap555/choudhury_18cse059/main/indian_food.csv"
data=pd.read_csv(path)
data.head()
```

```
data.dropna(axis=0,inplace=True)
```
